```python
import seaborn as sns
import pandas as pd
import seaborn as sns
import plotly.plotly as py
import plotly.graph_objs as go
import cufflinks as cf
import warnings; warnings.simplefilter('ignore')

%matplotlib inline

bank_data_full = pd.read_csv('bank-additional-full.csv')
```

```python
#Helper functions for transforming data, mainly months into appropriate months and years

def find_and_replace(df, month, prev, replace_with):
    for index, row in df.iterrows():
        if row['month'] == month:
            return
        elif row['month'] == prev:
            df.loc[index,'month'] = replace_with
```
