```python
import seaborn as sns
import pandas as pd
import seaborn as sns
%matplotlib inline

bank_data_full = pd.read_csv('bank-additional-full.csv')
```

```python
#First set of boxplots
```

```python
sns.mpl.rc("figure", figsize=(20,10))

#Boxplot education vs age
sns.boxplot(x = 'education', y = 'age', data = bank_data_full)
```

```python
#Boxplot outcome vs age

sns.boxplot(x = 'y', y = 'age', data = bank_data_full)
```

```python
#Boxplot job vs. age vs. outcome

sns.boxplot(x = 'job', y = 'age', hue = 'y',  data = bank_data_full)
```

```python
#Boxplot education vs. age vs. outcome

sns.boxplot(x = 'education', y = 'age', hue ='y', data = bank_data_full)
```

```python
#Jointplot number of employed vs. euribor3m

sns.jointplot(x = 'nr.employed', y = 'euribor3m', data = bank_data_full)
```

```python
sns.jointplot(x = 'emp.var.rate', y = 'euribor3m', data = bank_data_full)
```
